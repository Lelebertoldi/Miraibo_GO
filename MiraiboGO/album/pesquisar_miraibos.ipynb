```python
# Bibliotecas
from IPython.display import Image, display
import pandas as pd
import os
```

## Pesquisar os Miraibos

```python
# Lendo o arquivo CSV
df = pd.read_csv('miraibos.csv')

# Função para mostrar a imagem
def mostrar_imagem(caminho):
    display(Image(filename=caminho))

# Pergunta ao usuário se deseja pesquisar por ID ou nome
opcao = input("Você quer pesquisar por ID ou nome? (digite 'id' ou 'nome'): ").strip().lower()

if opcao == 'id':
    id_busca = input("Digite o ID que você deseja pesquisar: ")
    # Busca pela coluna ID, que agora é tratada como string
    resultado = df[df['id'].astype(str) == id_busca]
elif opcao == 'nome':
    nome_busca = input("Digite o nome que você deseja pesquisar: ")
    resultado = df[df['nome'].str.lower().str.contains(nome_busca.lower())]
else:
    print("Opção inválida.")
    resultado = None

# Exibindo a imagem se um resultado for encontrado
if resultado is not None and not resultado.empty:
    caminho_imagem = resultado.iloc[0]['imagem']
    print(f"Exibindo imagem para {resultado.iloc[0]['nome']}:")
    mostrar_imagem(caminho_imagem)
else:
    print("Nenhum resultado encontrado.")
```
